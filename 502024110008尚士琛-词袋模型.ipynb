```python
# 导入数据
import pandas as pd
```

```python
text_ana = pd.read_excel('text_analysis_twitter_sample.xlsx')
```

```python
text_ana
```

```python
text_ana = text_ana.drop(['Unnamed: 0'],axis=1)
```

```python
text_ana
```

```python
import re
def preprocess_text(text):
    # 转换大小写
    text = text.lower()
    
    # 移除URL
    text = re.sub(r'http\S+','',text)
    # 移除多余的空格
    text = re.sub(r'\s+', ' ',text).strip()
    return text
```

```python
text_ana['text'].apply(preprocess_text)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# 初始化CountVectorizer
vectorizer = CountVectorizer()
```

```python
# 将文本数据转化为词袋特征矩阵
X = vectorizer.fit_transform(text_ana["text"].apply(preprocess_text))
```

```python
# 展示特征矩阵的形状
print('词袋特征矩阵的形状:', X.shape)
```

原始数据500行7列 矩阵后500行文档4510是词的数量（行，列）

```python
# 展示词袋模型中的所有特征词
print('特征词列表:',vectorizer.get_feature_names_out()[:1000])
```

```python
len(vectorizer.get_feature_names_out())
```

```python
feature_words = vectorizer.get_feature_names_out()
```

```python
word_freq = dict(zip(feature_words, X.sum(axis=0).A1))
```
