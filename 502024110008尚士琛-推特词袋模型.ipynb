```python
import re
def preprocess_text(text):
    # 转换大小写
    text = text.lower()
    
    # 移除URL
    text = re.sub(r'http\S+','',text)
    # 移除特殊字符和数字
    text = re.sub(r'[^a-zA-Z\s]','',text)
    # 移除多余的空格
    text = re.sub(r'\s+', ' ',text).strip()
    # 移除停用词
    stop_words = set(['to','the','com','of','in','for','and','with','is','on','by','our','you','we','it','are','amp','rt','at','qt','this','my','be','from','as','have','that','&amp'])
    # print(text)
    text = ' '.join([word for word in text.split() if word not in stop_words])
    return text
```

```python
text_ana['text'].apply(preprocess_text)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# 初始化CountVectorizer
vectorizer = CountVectorizer(preprocessor=preprocess_text,token_pattern = r"(?u)\S\S+")
```

```python
# 将文本数据转化为词袋特征矩阵
X = vectorizer.fit_transform(text_ana["text"])
```

```python
# 展示特征矩阵的形状
print('词袋特征矩阵的形状:', X.shape)
```

原始数据500行7列 矩阵后500行文档4510是词的数量（行，列）
